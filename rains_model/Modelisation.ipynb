```python
def run_func(func,param=None):
  f =func
  p = param
  if type(p).__name__ == 'NoneType':
      f ()
  elif type(p).__name__ == 'tuple':
    if len(p) == 0:
      f ()
    elif len(p) == 1:
      a = p
      f (a)
    elif len(p) == 2:
      a,b = p
      f (a,b)
    elif len(p) == 3:
      a,b,c = p
      f (a,b,c)
    elif len(p) == 4:
      a,b,c,d = p
      f (a,b,c,d)
    elif len(p) == 5:
      a,b,c,d,e = p
      f (a,b,c,d,e)
    elif len(p) == 6:
      a,b,c,d,e,f = p
      f (a,b,c,d,e,f)
  elif type(p).__name__ == '_1p': # tuple as parameter !
    f (p.p1) 
  else:
    f(p)
```

```python
from typing_extensions import ParamSpec

###
class Task:

  libelle = ''
  def __init__(self,*arg):
    if len(arg) == 1:
      Task.libelle = self.libelle = arg[0]
    self.ta = self.task_ = {}
    self.tpa = self.task_['param'] = []
    self.tbe = self.task_['before'] = []
    self.taf = self.task_['after'] = []
    self.tap = self.task_['apply'] = []
    self.lib = self.task_['libelle'] = Task.libelle
    self.tta = self.tree_tasks = None

  def append_param(self,par):
    self.get()['param'].append( par )
    return self

  def push_param(self,branch:list,param):
    if self.tree_tasks == None:
      self.tta = Task.construct_tree_tasks(self)     
    leaf = self.tta
    for b in branch:
      leaf = leaf[b]
    if  type(leaf).__name__ != 'function':
      #lprint('LEAF ',leaf['task'].get()['param'])
      leaf['task'].get()['param'].append( param )
    else:
      self.get()['param'].append ( param )
    return self

  #def push_param(self,param):
  #  self.get()['param'].append ( param )

  def push_task_before_first(self,task,rb_tta:bool=True):
    self.get()['before'].insert(0,task)
    if rb_tta:
      self.tta = Task.construct_tree_tasks(self)
    return self

  def push_task_before_end(self,task,rb_tta:bool=True):
    self.get()['before'].append(task)
    if rb_tta:
      self.tta = Task.construct_tree_tasks(self)
    return self

  def push_task_apply_first(self,task,rb_tta:bool=True):
    self.get()['apply'].insert(0,task)
    if rb_tta:
      self.tta = Task.construct_tree_tasks(self)
    return self

  def push_task_apply_end(self,task,rb_tta:bool=True):
    self.get()['apply'].append(task)
    if rb_tta:
      self.tta = Task.construct_tree_tasks(self)
    return self

  def push_task_after_first(self,task,rb_tta:bool=True):
    self.get()['after'].insert(0,task)
    if rb_tta:
      self.tta = Task.construct_tree_tasks(self)
    return self

  def push_task_after_end(self,task,rb_tta:bool=True):
    self.get()['after'].append(task)
    if rb_tta:
      self.tta = Task.construct_tree_tasks(self)
    return self

  def get(self):
    return self.task_
  

  def apply_(self):
    param = None
    for be in self.task_['before']:
      be.apply_()
    for ap in  self.task_['apply']:
      if type( ap ).__name__ == 'function':
        if len(self.task_['param']) == 0:
          param = None
        else:
          param = self.task_['param'].pop(0)
        try:
          run_func( ap ,param)
        except Exception as ex:
          try:
            val = Default_param_mgnt.get_value_from_task_libelle(self.libelle)
            ap (val)
          except KeyError as kex:
            error(ex,ap,param)
      else:
        ap.apply_()
    for af in self.task_['after']:
      af.apply_()
    return self


  def log_in(self):
    lprint('>><< class ',self.libelle)


  def construct_tree_tasks(task,tt:OrderedDict()=None):
      if tt == None:
        tt = OrderedDict()

      tt['task'] = task

      ttbe = tt['before'] = OrderedDict()
      for t in task.tbe:
        ttbeli = ttbe[t.libelle] = OrderedDict()
        Task.construct_tree_tasks(t,ttbeli)

      ttap = tt['apply'] = OrderedDict()
      for t in task.tap:
        if type(t).__name__ != 'function':
          ttapli = ttap[t.libelle] = OrderedDict()
          Task.construct_tree_tasks(t,ttapli)
        else:
          ttapfu = ttap[t.__name__] = t

      ttaf = tt['after'] = OrderedDict()
      for t in task.taf:
        ttafli = ttaf[t.libelle] = OrderedDict()
        Task.construct_tree_tasks(t,ttafli)

      return tt
  
```

```python


def apply_keep_filtre(var:str,ope:str,val):

  global g_
  log_in()
  if ope == '==':
    g_['current_working_base'] = g_['current_working_base'].loc[g_['current_working_base'][var] == val]
  elif ope == '!=':
    g_['current_working_base'] = g_['current_working_base'].loc[g_['current_working_base'][var] != val]
  elif ope == '>':
    g_['current_working_base'] = g_['current_working_base'].loc[g_['current_working_base'][var] > val]
  elif ope == '<':
    g_['current_working_base'] = g_['current_working_base'].loc[g_['current_working_base'][var] < val]
  elif ope == '<=':
    g_['current_working_base'] = g_['current_working_base'].loc[g_['current_working_base'][var] <= val]
  elif ope == '>=':
    g_['current_working_base'] = g_['current_working_base'].loc[g_['current_working_base'][var] >= val]
  log_out()



def keep_(vars:list = []):

  global g_
  log_in()
  print_values (func='>keep',vars=vars,shape=g_['current_working_base'].shape)

  if len(vars) != 0:
    g_['current_working_base'] =  g_['current_working_base'][vars]

  print_values (func='<keep',vars=vars,shape=g_['current_working_base'].shape)
  log_out()


def filter_(filter:dict = {}):

  global g_
  log_in()
  print_values (func='>filtre',filter=filter,shape=g_['current_working_base'].shape)

  if len(filter) != 0:
    for var_,filtre_ in filter.items():
      ope,val = str_2_ope_val(filtre_)
      ope.strip()
      var_.strip()
      apply_keep_filtre(var_,ope,val)

  print_values (func='<filtre',filter=filter,shape=g_['current_working_base'].shape)
  log_out()


def drop_(vars:list = []):

  global g_
  log_in()
  print_values (func='>drop',vars=vars,shape=g_['current_working_base'].shape)
  
  if len(vars) != 0:
    for c in vars:
      g_['current_working_base'].drop(c,axis = 1, inplace=True)

  print_values (func='<drop',vars=vars,shape=g_['current_working_base'].shape)
  log_out()

```

```python

class Filtre(Task):
  def __init__(self):
    super().__init__()


class Keep(Filtre):
  libelle = 'Keep'
  def __init__(self):
    super().__init__()
    self.libelle = Keep.libelle
    self.get()['apply'].append( keep_ )


class Filter(Filtre):
  libelle = 'Filter'
  def __init__(self):
    super().__init__()
    self.libelle = Filter.libelle
    self.get()['apply'].append( filter_ )

class Drop(Filtre):
  libelle = 'Drop'
  def __init__(self):
    super().__init__()
    self.libelle = Drop.libelle
    self.get()['apply'].append( drop_ )

```

```python

def run_jobs(first_split:list):

  global g_
  g_['jobs'][g_['tag_job']] = []
  log_in()

  for ele_first in first_split:

    if len(g_['run_jobs_options']['only_ele_first']) != 0 and ele_first not in g_['run_jobs_options']['only_ele_first']:
      continue

    job = get_populed_job(ele_first)
    g_['jobs'][g_['tag_job']].append(job)

    for DF_type ,df in g_['projet']['part']['data']['jobs'][g_['tag_job']][ele_first]['dataframes']['second_split'].items():

      if len(g_['run_jobs_options']['only_DF_type']) != 0 and DF_type not in g_['run_jobs_options']['only_DF_type']:
        continue

      lprint ('DF_type: ',DF_type)
      for class_met in select_methode(DF_type):

        lprint ('DF',df)
        job.get()['tasks'].append( 
            get_populed_methode( 
                class_met,
                DF_type,
                df,
                job.get()['result']
            )
        )

    job.do_tasks()
  log_out()
```

## <font color='cyan'>Variables Categorielles</font>

La méthode suivante permet de visualiser la relation entre une variable categorielle et la probabilité de "RainTomorrow"

   

---

categorial_plot(
* location:str('')
* variable_cat:str
* set_tuple:tuple -> tuple décrivant l'ordonnancement des valeurs possibles de variable_cat lors de l'affichage
* color:str)

```python
%matplotlib inline

import matplotlib.pyplot as plt

#plt.use("Qt5Agg")

def move_figure(f, x, y):
    """Move figure's upper left corner to pixel (x, y)"""
    backend = plt.get_backend()
    if backend == 'TkAgg':
        f.canvas.manager.window.wm_geometry("+%d+%d" % (x, y))
    elif backend == 'WXAgg':
        f.canvas.manager.window.SetPosition((x, y))
    else:
        # This works for QT and GTK
        # You can also use window.setGeometry
        f.canvas.manager.window.move(x, y)

def categorial_plot(location:str,variable_cat:str,set_tuple:tuple,color:str):

  if loc == '':
    loc_df = copy.deepcopy(pd.DataFrame(rainData))
  else:
    loc_df = copy.deepcopy(rainData.loc[rainData['Location'] == location])

  loc_df = loc_df.dropna(axis = 0, how = 'any' , subset = [variable_cat,'RainTomorrow'])

  loc_df = loc_df.replace(to_replace =['Yes','No'], value=[1,0])
  loc_df = loc_df.loc[loc_df['RainToday'] == 0]
  #print (loc_df.RainTomorrow.count())
  #print (loc_df.RainTomorrow.sum())

  sum_isrt = {}
  nb_record = {}
  prob = {}

  for ele in set_tuple:
    sum_isrt[ele] = 0
    nb_record[ele] = 0
    prob[ele] = 0

  for i in loc_df.index:
    sum_isrt[loc_df[variable_cat][i]] += loc_df['RainTomorrow'][i]
    nb_record[loc_df[variable_cat][i]] += 1

  for k in prob:
    if nb_record[k] != 0:
      prob[k] = sum_isrt[k] / nb_record[k]
      prob[k] = np.round( prob[k],2)

  #print (sum_isrt)
  #print (nb_record)
  #print (prob)

  plt.bar(prob.keys(),prob.values(),label=location+' / '+variable_cat,color = color)
  plt.axhline(loc_df.RainTomorrow.sum()/loc_df.RainTomorrow.count())
  plt.legend();
  #f, ax = plt.subplots()
  #move_figure(f, 500, 500)
  plt.show()

```

Lorsque nous n'appliquons aucun filtres sur la localisation, nous constatons que certaines tendances apparaissent:

*   Concernant les vents, aucune direction particulière ne semble être un indicateur de pluie; ceci est en cohérence avec nos observations: l'influence des directions de vents est totalement lié aux localités. par contre, il semble que une vitesse de vent élévée et en particulier des bourrasques, sont des indicateurs de pluie à venir.
*   Concernant les autres variables, il semble que l'on puisse observer certaines tendances partagées quelquesoit la localisation
 *   Des temperatures basses l'après-midi
 *   Un taux d'humidité élévé
 *   Une couverture nuageuse
 *   Un taux d humidité élévé
 *   Des pressions basses

 sont généralement des indicateurs de pluie à venir
 
Il est, par ailleurs important de considerer que les localisations implique de s fourchettes de valeurs pour une majorité de variables; par exemple les temperatures min et max ne sont les mêmes à Sydney et à Perth ...

Ces considérations plaident en la nécéssité d'appliquer les analyses et traitements par localité.

Dernier point à considérer: Est il judicieux de chercher à prévoir la pluie pour le lendemain, alors qu'il pleut déjà aujourd'hui (variable 'Raintoday') ?

Objectivement nous pensons que non, d'autant que les conditions climatiques d'une journée de pluie et de beaux temps (vents, humidité, temp) sont très différentes) et donc non mixable pour l'analyse.

En d'autres termes, quels sont les changements climatiques, un jour de temps sec, indiquant une probabilité élevé de pluie le lendemain.

   *  les variables une par une (categorielle ou numérique)
   *  les variables catégorielles une par une (numérisées)
